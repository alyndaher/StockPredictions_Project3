```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
sector_csv = 'data/tensectors_final.csv'

sector_df = pd.read_csv(sector_csv)
```

```python
#drop unnamed column and two statistically not-significant variables from previous model
sector_df = sector_df.drop(columns=['Unnamed: 0','Real Estate','Telecommunication Services'],axis=1)
```

```python
sector_df
```

```python
#sklearn output for intercept and coefficients - removed Real Estate & Telecommunication Services

X = sector_df[['Consumer Staples','Information Technology','Ameriprise Financial','Consumer Discretionary','Utilities','Health Care','Materials','Energy','Industrials']] #8 variables for multiple regression
y = sector_df['S&P']
 
# with sklearn fit the model with all of x dependent variables or features
regr = linear_model.LinearRegression()
regr.fit(X, y)

model = LinearRegression()
model.fit(X,y)
score = model.score(X,y)

print('Intercept: ', regr.intercept_)
print('Coefficients: ', regr.coef_)
print('R2 Score: ', score)
```

```python
# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```

```python
#Create a prediction with sklearn for Aug 3, 2020

con_staples = 129.3
info_tech = 435.75
financial = 155.69
con_disc = 3111.89
utilities = 147.06
health = 152.84
materials = 13.68
energy = 5.52
industrials = 31.88


print ('Predicted S&P500: ', regr.predict([[con_staples, info_tech, financial, con_disc, utilities, health, materials, energy, industrials]]))
```
